### problema 44 : Sala Cine
 Necesitamos un sistema básico para organizar la información y las tareas de un cine. Para lograrlo, usaremos la Programación Orientada a Objetos (POO) para crear un "molde" digital de una sala de cine.
 ![link text](https://raw.githubusercontent.com/RamirezAlvarezazul128/Act-24-clase-SalaCine-Ramirez-Azul-0441-/refs/heads/main/Act%2024%20Ramirez%20Azul%200441.png)

```python
# Lenguaje: Python - Nivel: Principiante
# Clase: SalaCine

class SalaCine:
    # 1. Atributos: numeroSala, capacidadButacas, tipoPantalla, sistemaSonido
    def __init__(self, numeroSala, capacidadButacas, tipoPantalla, sistemaSonido):
        self.numeroSala = numeroSala
        self.capacidadButacas = capacidadButacas
        self.tipoPantalla = tipoPantalla
        self.sistemaSonido = sistemaSonido

    # 2. Función: proyectarPelicula()
    def proyectarPelicula(self):
        print(f"🎬 ACCIÓN: Proyectando película en la Sala {self.numeroSala}.")
        print(f"   Detalles: Pantalla {self.tipoPantalla} con sonido {self.sistemaSonido}.")

    # 3. Función: ajustarIluminacion()
    def ajustarIluminacion(self):
        print(f"💡 Luces reguladas para recibir a {self.capacidadButacas} personas.")

    # 4. Función: limpiarButacas()
    def limpiarButacas(self):
        print(f"🧹 Iniciando limpieza de butacas en la Sala {self.numeroSala}...")

# Función para interactuar con el usuario y pedir datos
def pedir_datos(nombre_objeto):
    print(f"\n--- Ingrese los atributos para {nombre_objeto} ---")
    num = input("Número de Sala: ")
    cap = input("Capacidad de Butacas: ")
    pan = input("Tipo de Pantalla: ")
    son = input("Sistema de Sonido: ")
    return SalaCine(num, cap, pan, son)

# --- Creación de los 2 objetos solicitados ---
sala1 = pedir_datos("sala1")
sala2 = pedir_datos("sala2")

# --- Ejecución de las actividades (Funciones) ---
print("\n" + "="*45)
print("REPORTE DE ACTIVIDADES DEL CINE")
print("="*45)

# Actividades para el objeto sala1
print(f"\n>>> PROCESANDO SALA: {sala1.numeroSala}")
sala1.ajustarIluminacion() # Esta es la función que mencionabas
sala1.proyectarPelicula()
sala1.limpiarButacas()

print("-" * 20) # Separador visual

# Actividades para el objeto sala2
print(f">>> PROCESANDO SALA: {sala2.numeroSala}")
sala2.ajustarIluminacion() # Verificado: Ahora aparecerá claramente para la sala 2
sala2.proyectarPelicula()
sala2.limpiarButacas()

print("\n" + "="*45)
print("fin del programa")
print("programa realizado por Ramirez Azul NL = 44")
```
